```python
import pandas as pd
import numpy as np
import os
from re import search
import codecs
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import missingno as msno
```

```python
combined_data = pd.read_csv('combined_data.csv',encoding = 'unicode_escape')
```

```python
combined_data = combined_data.replace('?',None)
combined_data = combined_data.replace('nan',None)
```

```python
combined_data['ID'] = combined_data['ID'].astype(str)
combined_data['Start_date'] = pd.to_datetime(combined_data['Start_date'])
combined_data['market_cap'] = combined_data['market_cap'].astype(str)
combined_data['Sector'] = combined_data['Sector'].astype(str)
combined_data['CEO'] = combined_data['CEO'].astype(str)
combined_data['No.of employee'] = combined_data['No.of employee'].fillna(0)
combined_data['No.of employee'] = combined_data['No.of employee'].astype(int)
combined_data['Revenue (Cr)'] = combined_data['No.of employee'].astype(int)
combined_data['Net-valuation (Cr)'] = combined_data['No.of employee'].astype(int)
combined_data['Share_price (RS)'] = combined_data['Share_price (RS)'].astype(float)
combined_data['Type'] = combined_data['Type'].astype(str)
combined_data['Product'] = combined_data['Product'].astype(str)
combined_data['industry'] = combined_data['industry'].astype(str)
combined_data['Company_background'] = combined_data['Company_background'].astype(str)
combined_data['product_info'] = combined_data['product_info'].astype(str)
```
