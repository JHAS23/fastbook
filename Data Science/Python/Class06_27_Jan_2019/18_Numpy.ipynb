# Filtering values from an Array

```python
# Syntax:
#     np.where(<condition>)
#     np.where((<condition1>) & (<condition2..>))
#     will return indices of row and col
```

```python
tmp3 = np.random.randint(1,90,100).reshape(10,10)
tmp3
```

```python
np.where(tmp3>30)
```

```python
tmp3[np.where(tmp3>30)]
```

```python
tmp3[np.where(tmp3>30)].size
```

# Create a square mtarix of 10,10 and make all the upper diagonal values to zero and same thing with lower triangle

```python
tmp4 = np.random.randint(1,500,100).reshape(10,10)
tmp4
```

# For Upper triangle

```python
cnt = 0
for i in range(len(tmp4)):
    tmp = tmp4[i]
    tmp[cnt+1:] = 0
    cnt += 1
tmp4
```

# For Lower triangle

```python
tmp5 = np.random.randint(1,500,100).reshape(10,10)
cnt = 0
for i in range(len(tmp5)):
    tmp = tmp5[i]
    tmp[:cnt] = 0
    cnt += 1
tmp5
```

```python
tmp5 = np.random.randint(1,500,100).reshape(10,10)
cnt = 0
for row in (tmp5):
    row[:cnt] = 0
    cnt += 1
tmp5
```
