```python
a = 2
b = 0
print(a/b)
print("I am after a/b")
```

```python
# Syntax:
#     try:
#         <write your doubtful code which may produce exception>
#     except:
#         <write other part of code>
#     finally: # <optinal block, will be exacuted irrespective of exception comes or not>
        
```

```python
a = 2
b = int(input("Enter a number..."))

try:
    print(a/b)
except:
    print("I am in except block")
```

```python
a = 2
# b = int(input("Enter a number..."))
try:
    b = int(input("Enter a number..."))
    print(a+b)
except Exception as err: # err is just a variable which will catch meaningfull message of exception
    print(err)
```

# Using Finnaly

```python
a = 2
# b = int(input("Enter a number..."))
try:
    b = int(input("Enter a number..."))
    print(a+b)
except Exception as err: # err is just a variable which will catch meaningfull message of exception
    print(err)
finally:
    print("I am in finally block")
```
