# For loop

```python
# for <variable> in <sequence>:
#     <statements>
#     .
#     .
```

```python
a = [1,2,3,4,5]
```

```python
print(a[0]**2)
print(a[1]**2)
print(a[2]**2)
print(a[3]**2)
print(a[4]**2)
```

```python
for tmp in a:
    print(tmp**2)
```

# While loop

```python
# while <condition>:
#     <statements>
#     .
#     .
```

```python
a = 10
while a >0:
    print(a)
    a -= 1
    # or a = a-1
```

# Break

```python
tmp2 = "python"

for char in tmp2:
    if char == "h":
        break
    else:
        print(char)
```

# Continue

```python
tmp2 = "python"

for char in tmp2:
    if char == "h":
        continue
    else:
        print(char)
```

# Generate a list of even and odd numbers between 1 to 100 and print  them as a separate list of "Even" & "Odd"

```python
tmp4 = range(1,100)
```

```python
Even = []
Odd = []

for itm in tmp4:
    if itm%2 == 0:
        Even.append(itm)
    else:
        Odd.append(itm)
print("Even list::",Even)
print("\nOdd list::",Odd)
```
