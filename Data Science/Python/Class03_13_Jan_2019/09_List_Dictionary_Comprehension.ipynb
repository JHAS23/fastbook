# List Comprehension

```python
# Syntax: [<variable> for <variable> in <sequence>]
# Syntax: [<variable> for <variable> in <sequence> <if condition optional>]
```

```python
print([var for var in range(1,100) if var%2!=0])
```

```python
print([var for var in range(1,100) if var%2==0])
```

# Dictionary Comprehension

```python
# Syntax: {<variable>:<variable> for <variable> in <sequence>}
# Syntax: {<variable>:<variable> for <variable> in <sequence> <if condition optional>}
```

#### Create a dictionary using dict_comprehansion in which Key is an individual number and their value will be power of 3 of that key . Number range will be 1 to 20

```python
print({var:var**3 for var in range(1,20)})
```

```python
print({var:var**3 for var in range(1,20) if var>4 and var<15})
```
