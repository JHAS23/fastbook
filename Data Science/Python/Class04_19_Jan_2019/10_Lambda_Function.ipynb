# Lambda an anonymous function

```python
# Syntax: 
#     lambda <var1>,<var2>,....:<expression>
#     lambda <var1>,<var2>,....:<expression> if <condition> else <expression>
```

```python
lam = lambda x: x**2
```

```python
lam(2)
```

```python
lam2 = lambda x,y: x+y
```

```python
lam2(1,2)
```

```python
lam3 = lambda x,y=10: x+y
```

```python
lam3(1)
```

```python
lam3(1,20)
```

# lambda with condition

```python
lam4 = lambda x: x**2 if x%2==0 else x**3
```

```python
lam4(3)
```

```python
[lam4(var) for var in range(100)]
```
