# Regular Expressions

```python
import re
```

```python
# match:
#     re.match("<pattern>",<searchable string>,<flags>)
# search:
#     re.search("<pattern>",<searchable string>,<flags>)    
```

```python
tmp = "This is Learnbay class5"
```

```python
# mObj = re.match("Learnbay",tmp)
mObj = re.match("(This) (is)",tmp)
print(mObj)
```

```python
if mObj:
#     print(mObj.group())
#     print(mObj.group(2))
    print(mObj.groups())
    
```

# Search

```python
tmp = "This is Learnbay class5"
```

```python
# sObj = re.search("Learnbay",tmp)
# sObj = re.search("Learnbay (class5)",tmp)
# sObj = re.search("LeaRNbay (Class5)",tmp)
sObj = re.search("LeaRNbay (Class5)",tmp,re.I)
# print(sObj)

if sObj:
    print(sObj.group())
#     print(sObj.group(1))
#     print(sObj.groups())
    
```

```python
tmp2 = "This is Learn\nbay class5"
```

```python
sObj = re.search("LEArn.bay",tmp2,re.I|re.S) # re.S or re.DOTALL, all the flags
                                            # must be seperated by "|"
print(sObj)

if sObj:
    print(sObj.group())
#     print(sObj.group(1))
#     print(sObj.groups())
    
```

# Substitute

```python
tmp3 = "This is learnbay class class class class class"
# new_string = re.sub("class","CLASS",tmp3)
new_string = re.sub("class","CLASS",tmp3,3)
print(new_string)
```

```python
tmp4 = "12345ABC^@#%^$#%"
```

```python
# sObj = re.search("\d+",tmp4)
# sObj = re.search("[0-9]+",tmp4)
# sObj = re.search("\w+",tmp4)
# sObj = re.search("[a-zA-Z]+",tmp4)
# sObj = re.search("[aA-zZ]+",tmp4)
# sObj = re.search("\W+",tmp4)
sObj = re.search("(\d+)([a-zA-Z]+)(\W+)",tmp4)

if sObj:
#     print(sObj.group())
    grp = sObj.groups()
    for itm in grp:
        print(itm)
```

```python
tmp5 = "1111111 222 ABC 345 1111 DEF"
```

```python
# sObj = re.search("(.*)?([a-zA-Z]+)(.*)",tmp5)
sObj = re.search("([a-zA-Z]+)(.*)",tmp5)
print(sObj)

if sObj:
    print(sObj.group(1))
    print(sObj.group(2))
#     print(sObj.group(3))
```

# Back referencing 

```python
tmp6 = "111 111 111 222 111 111 222 111 "
# sObj = re.search("111 111 111 ",tmp6)
sObj = re.search("(111 )(\\1)(\\2)(222 )\\1\\1\\4",tmp6)

if sObj:
    print(sObj.group())
```

```python
tmp6 = "111 111 111 222 111 111 222 111 "
# sObj = re.search("111 111 111 ",tmp6)
sObj = re.search("111",tmp6)

if sObj:
    print(sObj.group())
```
