# Seek function 
- to modify the pointer position

```python
# seek(<offset>,<pointer position>)
# where pointer position: 
#     0(start of the file <always set to default>),
#     1(current position),
#     2(from end of the file)
```

```python
fh = open("test1.txt")
print(fh.read(2))
fh.seek(4)
print(fh.read(2))
fh.seek(0)
print(fh.read(2))
```

```python
fh = open("test1.txt","rb")
print(fh.read(2))
fh.seek(4,1)
print(fh.read(2))
```
