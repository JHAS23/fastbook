# Plot using MatplotLib

```python
tmp4 = np.random.randint(1,50,100)
tmp4
```

# Line Plot

```python
plt.figure(figsize=(16,4))
plt.plot(range(100),tmp4)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.bar(range(100),tmp4)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.barh(range(100),tmp4)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.scatter(range(100),tmp4)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.scatter(range(100),tmp4)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter-Plot")
plt.xticks(range(0,100,5),rotation=90)
plt.yticks(range(0,50,5))
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(100):
    plt.scatter(idx,tmp4[idx],s=.2*np.pi*tmp4[idx]**2,alpha=.3)
    if tmp4[idx] >=48:
        plt.scatter(idx,tmp4[idx],s=.8*np.pi*tmp4[idx]**2,alpha=.7,marker="<")    
plt.show()
```

```python
x = np.arange(100)
y1= np.random.randint(1,100,100)
y2= np.random.randint(1,100,100)
y3 = np.random.randint(1,100,100)
plt.figure(figsize=(16,4))
plt.bar(x[:20],y1[:20],width=.8)
plt.bar(x[:20],y2[:20],width=.8,bottom=y1[:20])
plt.bar(x[:20],y3[:20],width=.8,bottom=y1[:20]+y2[:20])
plt.show()
```

```python
plt.figure(figsize=(5,5))
y = [5,7,3,8,3,9,3,1]
plt.pie(y,shadow=True,labels=y,labeldistance=.8,center=(1,2),explode=[.9,0,0,.1,0,0,.1,0])
plt.show()
```
